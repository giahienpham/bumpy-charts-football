```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplsoccer
import highlight_text
from mplsoccer import Bumpy
from PIL import Image
from io import BytesIO
import requests
```

```python
df = pd.read_csv('La Liga.csv')
```

```python
df.head()
```

```python
df = df.T
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
df.head()
```

```python
week = ['Week ' + str(num) for num in range (1, 35)]
```

```python
highlight_dict = {
    'Atletico Madrid' : '#CB3524',
    'Real Madrid': 'white',
    'Barcelona': '#EDBB00',
    'Sevilla':'green',
}
```

```python
bumpy = Bumpy(
    scatter_color = '#282828',
    line_color = '#252525',
    rotate_xticks = 90,
    ticklabel_size = 15,
    scatter_primary = 'D',
    show_right = True,
    plot_labels = True,
    alignment_yvalue = .1,
    alignment_xvalue = .065 
)
```

```python
fig, ax = bumpy.plot(
    x_list = week,
    y_list = np.linspace(1,20,20).astype(int),
    values = df,
    secondary_alpha= .5,
    highlight_dict = highlight_dict,
    figsize = (18, 18),
    # y_label = 'Position',
    ylim = (-.1,23),
    lw = 2.5
)

ax.set_ylabel('Position', labelpad=20) 

fig.text(s= 'La Liga through week 34', x = .17, y = .84,
         c = 'white', size = 24, weight = 'bold'
         )
highlight_text.fig_text(x=.17, y = .82,
                        s = 'Comparing <Atletico>, <Real Madrid>, <Barcelona> & <Sevilla>', highlight_textprops = [
                            {'color': '#CB3524'},
                            {'color': 'white'},
                            {'color': '#EDBB00'},
                            {'color': 'green'}
                        ], color = 'white', size = 16)

ax2 = fig.add_axes([.05,.77,.1,.2])
ax2.axis('off')
img = Image.open('logo.jpg')
ax2.imshow(img)
```
